```python
rename_cols = {
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_c',
    't2':'temp_feels_like_c',
    'hum':'humidity_percentage',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

bikes.rename(rename_cols, axis=1, inplace=True)
```

```python
bikes.head()
```

```python
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}
```

```python
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}
```

```python
#changing datatype 
bikes.season = bikes.season.astype('str')
```

```python
#mapping seasons to dictionary
bikes.season = bikes.season.map(season_dict)
```

```python
#changing datatype 
bikes.weather = bikes.weather.astype('str')
```

```python
#mapping weather to weather dictionary
bikes.weather = bikes.weather.map(weather_dict)
```
